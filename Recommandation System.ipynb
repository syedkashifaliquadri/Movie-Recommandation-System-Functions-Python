# Movie Recommendation System

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('u.data',sep='\t',header=None, names = ['user_id','item_id','rating','timestamp'])
df1.head()
```

```python
df2=pd.read_csv('Movie_Id_Titles')
df2.head()
```

```python
df1=pd.merge(df1,df2,on='item_id')
df1.head()
```

```python
def get_top_rated(df1, thr=80, no=10):
    df3=df1.groupby('title').rating.mean()
    df4=pd.DataFrame(df3)
    df4['no.']=df1.groupby('title').rating.count()
    df5=df4[df4['no.']>thr]
    return df5.sort_values(['rating'], ascending=[0]).head(no)
get_top_rated(df1, 80, 10)
```

```python
def get_most_reviewed_popu(df1, no=10):
    df3=df1.groupby('title').rating.sum()
    df6=pd.DataFrame(df3)
    df6['no.']=df1.groupby('title').rating.count()
    maxno=df6['no.'].max()
    df6['rating'] = df6['rating']/maxno
    return df6.sort_values(['rating'], ascending=[0]).head(no)
get_most_reviewed_popu(df1)
```

```python
def get_curated(df1, thr=80, no=10):
    df3=df1.groupby('title').rating.mean()
    df7=pd.DataFrame(df3)
    df7['no.']=df1.groupby('title').rating.count()
    df8=df7[df7['no.']>thr]
    maxno=df8['no.'].max()
    #use 0.5 for Adjusting rating
    
    df8['rating'] += df8['no.']/maxno - 0.5
    return df8.sort_values(['rating'],ascending=[0]).head(no)
get_curated(df1)
```

```python
movie_matrix = df1.pivot_table(index='user_id', columns='title', values='rating')
movie_matrix.head()
```

```python
movie_matrix.shape
```

```python
def get_similar(movie_matrix, movie_name, thr=40, no=10):
    SW_user_rating = movie_matrix[movie_name]
    similar_to_sw = movie_matrix.corrwith(SW_user_rating) # ("same user star wars k elawa kis kis 
                                                                       # movie mai rating de raha hai")
    corr_SW = pd.DataFrame(similar_to_sw, columns=['correlation'])
    corr_SW.dropna(inplace=True)
    dftemp = df1.groupby('title').rating.count()
    dftemp = pd.DataFrame(dftemp)
    corr_SW = corr_SW.merge(dftemp, on='title')
    corr_SW = corr_SW[corr_SW['rating']>thr]
    
    return corr_SW.sort_values(by='correlation', ascending=False).head(no)

get_similar(movie_matrix, 'Star Wars (1977)')
```

```python
#Yaha hum 2 users k similar decision nikale gye or pher batayge k kn se users same decision lete hai.
def get_sim_users_reviews(df1, movie_matrix, uid, thr=20):
    SW_user_rating = movie_matrix.loc[uid]
    similar_to_sw = movie_matrix.corrwith(SW_user_rating, axis=1)
    similar_to_sw.head()
    corr_SW = pd.DataFrame(similar_to_sw, columns=['correlated'])
    corr_SW.dropna(inplace=True)
    corr_SW.head()
    coUsers = df1.groupby('user_id').rating.count()
    coUsers = pd.DataFrame(coUsers)
    coUsers.columns = ['no.']
    coUsers.head()
    
    corr_SW = corr_SW.join(coUsers['no.'])
    
    corr_SW = corr_SW[corr_SW['no.']>thr]
    corr_SW = corr_SW.sort_values(by = 'correlated', ascending = False)
    corr_SW = corr_SW[corr_SW['correlated']>0.5]
    corr_SW = corr_SW.sort_values(by = 'correlated', ascending = False)
    
    
    
    uids = corr_SW.index
    newDF = df1[df1.user_id.isin(uids)]
    
    return newDF




def get_sim_user_top_rated(df1, movie_matrix, uid, thr=20):
    newDF =get_sim_users_reviews(df1, movie_matrix, uid, thr=20)
    return get_top_rated(newDF,10,10)
```
